```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/retail_sales_dataset.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract month and year for seasonal analysis
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# --- Seasonal Sales Trends ---
# Aggregate sales data by year and month
monthly_sales = data.groupby(['Year', 'Month'])['Total Amount'].sum().reset_index()

# Plot seasonal sales trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Total Amount', hue='Year', marker='o')
plt.title('Seasonal Sales Trends', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales Amount', fontsize=12)
plt.xticks(range(1, 13))
plt.legend(title='Year')
plt.grid(True)
plt.show()

# --- Top Product Categories by Revenue ---
# Aggregate revenue by product category
product_revenue = data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

# Plot top product categories by revenue
plt.figure(figsize=(10, 6))
sns.barplot(x=product_revenue.values, y=product_revenue.index, palette='viridis')
plt.title('Top Product Categories by Revenue', fontsize=16)
plt.xlabel('Total Revenue', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
plt.grid(axis='x')
plt.show()

# --- Revenue Trend Over Time ---
# Aggregate revenue by date
daily_revenue = data.groupby('Date')['Total Amount'].sum().reset_index()

# Plot revenue trend over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_revenue, x='Date', y='Total Amount', color='blue', marker='o')
plt.title('Revenue Trend Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.grid(True)
plt.show()

```
